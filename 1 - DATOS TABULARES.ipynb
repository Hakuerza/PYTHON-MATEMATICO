Vamos a ver cómo importar nuestro primer módulo: [numpy](https://numpy.org/doc/stable/reference/index.html). Numpy es el módulo numérico de python y nos permite aplicar operaciones matemáticas sobre colecciones de datos llamadas arrays. Para el análisis de datos y el data science vectores y matrices son imprescindibles. Los tensores son comunes a la hora de utilizar técnicas de deep learning.

#### Arrays

```python
a1D = np.array([1, 2, 3, 4])
a2D = np.array([[1, 2], [3, 4]])
a3D = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
```

Podemos definir el tipo de dato

```python
a = np.array([127, 128, 129], dtype=np.int8)
```

```python
print(a)
```

Podemos hacer operaciones con los arrays

```python
a = np.array([2, 3, 4], dtype=np.int8)
b = np.array([5, 6, 7], dtype=np.int8)
c = a - b
d = a * b
e = np.dot(a,b)
```

```python
print(c)
print(b)
print(e)

print(d)

print(np.sqrt(a))
```

```python
df_titanic = pd.read_csv("1_modulo_matematico/titanic_train.csv")
```

```python
# con head vemos las primeras filas del dataframe
df_titanic.head()
```

```python
# podemos ver cuantas filas y columnas tiene el dataframe
df_titanic.shape
```

### Selección de campos y registros

```python
# utilizando el operador loc nos referimos a las filas y a las columnas por su NOMBRE:
df_titanic.loc[0, "Name"]
```

```python
df_titanic.loc[5, :]
```

```python
df_titanic.loc[:, "Name"]
```

```python
# La operacion loc resulta muy util para filtrar los registros:
df_titanic.loc[df_titanic["Age"] < 18]
```

```python
# Despues podemos ver el numero de registros
len(df_titanic.loc[df_titanic["Age"] < 18])
```

```python
# Podemos combinar las condiciones
len(df_titanic.loc[(df_titanic["Age"] < 18) & (df_titanic["Sex"] == "female")])
```

```python
# Seria equivalente excluyendo el .loc
df_titanic[(df_titanic["Age"] < 18) & (df_titanic["Sex"] == "female")].head()
```

```python
# Se pueden concatenar filtros
df_titanic[(df_titanic["Age"] < 18)][(df_titanic[(df_titanic["Age"] < 18)]["Sex"] == "female")].head()
```

```python
# Podemos crear una columna utilizando estas funciones
df_titanic["mayor_edad"] = 0
df_titanic.loc[df_titanic["Age"] >= 18, "mayor_edad"] = 1
df_titanic.head(15)
```

```python
# Podemos eliminar columnas
del df_titanic["mayor_edad"]
df_titanic.head()
```

```python
# Utilizando el operador .iloc nos referimos
df_titanic.iloc[:, 5]
```

### Estructura de la tabla

```python
# Podemos ver las columas del data frame:
print(df_titanic.columns)
print(df_titanic.columns.tolist())
```

### Contar rápidamente valores

```python
# Podemos ver los valores que tienen las distintas columnas
df_titanic["Sex"].value_counts()
```

### Agrupaciones
Una agrupación consiste en reducir la dimensión del dataframe quedándonos solo con los valores disponibles en ciertas columnas y generando columnas a partir de otras existentes y aplicando operaciones de agregación como contar, sumar, calcular el máximo, etc...

```python
df_titanic.groupby(["Sex", "Survived"]).agg({"Name": "count"})
```
